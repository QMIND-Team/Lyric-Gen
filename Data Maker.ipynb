```python
import lyricsgenius as lg
import re

token = 'ybsd5TsGZ-xJbwlbaQ6iHpB-b92qbdPJHlKYBgsr5x0Ugq7RdbDSrhb057ngOJ3S'
genius = lg.Genius(token)
```

```python
artist = 'Kanye West'
result = genius.search_artist(artist, max_songs=150, sort='popularity')
print(result.songs)
```

```python
import pandas as pd
df = pd.DataFrame(columns=['song','lyrics'])
df
```

```python
for row in range(len(result.songs)):
    song = result.songs[row].title
    lyrics = genius.search_song(song, artist).lyrics
    
    data = {'song': song, 'lyrics': lyrics}
    df = df.append(data, ignore_index=True)
```

```python
def clean(data):
    data = data.replace('\n\n', '\n')
    data = re.sub('\[([A-Za-z 0-A-Za-z_]+)\]\\n', '', data)
    
    return data

df['lyrics'] = df['lyrics'].apply(clean)
```

```python
df.to_csv('Data/' + artist.lower().replace(' ', '_') + '_dataset.csv')
```
